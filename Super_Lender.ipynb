```python
def handle_non_numerical_data(train_data2):
    columns = train_data2.columns.values
    for column in columns:
        text_digit_vals = {}
        def convert_to_int(val):
            return text_digit_vals[val]
        
        if train_data2[column].dtype != np.int64 and train_data2[column].dtype != np.float64:
            column_contents = train_data2[column].values.tolist()
            unique_elements = set(column_contents)
            x = 0
            for unique in unique_elements:
                if unique not in text_digit_vals:
                    text_digit_vals[unique] = x
                    x += 1
            train_data2[column] = list(map(convert_to_int, train_data2[column]))
    
    return train_data2

train_data2 = handle_non_numerical_data(train_data2)
train_data2.head()
```

```python
def handle_non_numerical_data(test_data2):
    columns = test_data2.columns.values
    for column in columns:
        text_digit_vals = {}
        def convert_to_int(val):
            return text_digit_vals[val]
        
        if test_data2[column].dtype != np.int64 and test_data2[column].dtype != np.float64:
            column_contents = test_data2[column].values.tolist()
            unique_elements = set(column_contents)
            x = 0
            for unique in unique_elements:
                if unique not in text_digit_vals:
                    text_digit_vals[unique] = x
                    x += 1
            test_data2[column] = list(map(convert_to_int, test_data2[column]))
    
    return test_data2

test_data2 = handle_non_numerical_data(test_data2)
test_data2.head()
```
